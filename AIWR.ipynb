## Creation of Inverted index:

```python
def create_inverted_index(tokenized_tweets):
    inverted_index = dict()
    for document_idx in range(len(tokenized_tweets)):
        for word in tokenized_tweets[document_idx]:
            if word not in inverted_index:
                inverted_index[word] = list()
            inverted_index[word].append(document_idx)
    
    for key in inverted_index:
        inverted_index[key] = list(set(inverted_index[key]))
        
    return inverted_index
```

## Handling wild card and phase queries:

### a) Boolean Model:

```python
def parse_query(infix_tokens):
    
    precedence = {}
    precedence['NOT'] = 3
    precedence['AND'] = 2
    precedence['OR'] = 1
    precedence['('] = 0
    precedence[')'] = 0    

    output = []
    operator_stack = []

    for token in infix_tokens:
        if (token == '('):
            operator_stack.append(token)

        elif (token == ')'):
            operator = operator_stack.pop()
            while operator != '(':
                output.append(operator)
                operator = operator_stack.pop()

        elif (token in precedence):
            if (operator_stack):
                current_operator = operator_stack[-1]
                while (operator_stack and precedence[current_operator] > precedence[token]):
                    output.append(operator_stack.pop())
                    if (operator_stack):
                        current_operator = operator_stack[-1]
            operator_stack.append(token) # add token to stack
        else:
            output.append(token.lower())

    while (operator_stack):
        output.append(operator_stack.pop())

    return output
```

```python
def boolean_query(query, inverted_index):
    query = query.strip()
    query_tokens = query.split()
    boolean_query = parse_query(query_tokens)
        
    result_stack = list()
    for idx, token in enumerate(boolean_query):
        if token not in ["AND", "NOT", "OR"]:
            result = set(inverted_index[token])
        else:
            if token in ['AND', 'OR']:
                right_operand = result_stack.pop()
                left_operand = result_stack.pop()
                
                if token == 'AND':
                    operation = set.intersection
                else:
                    operation = set.union
                
                result = operation(left_operand, right_operand)
                
            else:
                operand = result_stack.pop()
                complement_document_ids = inverted_index[boolean_query[idx-1]]
                result = list()
                for word in inverted_index:
                    result.extend([_id for _id in inverted_index[word] if _id not in complement_document_ids])
                result = set(result)
                
        result_stack.append(result)
    
    return result_stack.pop()
```

### b) Positional_indexing:

```python
def construct_positional_posting_list(tokenized_corpus):
    positional_index = dict()
    for tweet_id, tweet in enumerate(tokenized_corpus):
        for token_id, token in enumerate(tweet):
            if token not in positional_index:
                positional_index[token] = dict()
            if tweet_id not in positional_index[token]:
                positional_index[token][tweet_id] = list()
            positional_index[token][tweet_id].append(token_id)
    
    for token in positional_index:
        for tweet_id in positional_index[token]:
            positional_index[token][tweet_id] = sorted(positional_index[token][tweet_id])
        items = list(positional_index[token].items())
        items.sort(key=lambda x: x[0])
        for k, v in items:
            positional_index[token][k] = v
            
    return positional_index
```

```python
def positional_intersect(p1, p2, K):
    answer = list()
    i = 0
    j = 0
    while i < len(p1) and j < len(p2):
        document_id_p1 = list(p1.keys())[i]
        document_id_p2 = list(p2.keys())[j]
        
        if document_id_p1 == document_id_p2:
            l = list()
            pp1 = p1[document_id_p1]
            pp2 = p2[document_id_p2]
            
            k = 0
            while k < len(pp1):
                m = 0
                while m < len(pp2):
                    distance = pp2[m] - pp1[k]
                    if distance == K:
                        l.append(m)
                    m += 1
                
                for ps in l:
                    distance = (pp2[ps] - pp1[k])
                    if distance != K:
                        l.remove(ps)
                        
                for ps in l:
                    answer.append((document_id_p1, pp1[k], pp2[ps]))
                
                k += 1
            
            i += 1
            j += 1
        
        elif document_id_p1 < document_id_p2:
            i += 1
        else:
            j += 1
    
    return answer
```

## Retrieval and Ranking of retreived documents:

```python
import math
from collections import Counter

import re
from collections import Counter

def create_doc_vector(doc):
    # remove non-alphanumeric characters and convert to lowercase
    cleaned_doc = re.sub(r'\W+', ' ', doc.lower())
    # create a counter of the terms in the document
    term_counts = Counter(cleaned_doc.split())
    return term_counts

def create_query_vector(query_terms, documents):
    # Calculate the term frequency of each query term
    term_freq = {}
    for term in query_terms:
        term_freq[term] = term_freq.get(term, 0) + 1
    
    # Calculate the inverse document frequency of each query term
    num_docs = len(documents)
    epsilon = 1e-7  # Add a small constant to the denominator to avoid math domain error
    idf = {term: math.log((num_docs + epsilon) / count) for term, count in term_freq.items()}
    
    # Create the query vector by combining term frequency and inverse document frequency
    query_vector = {term: freq * idf[term] for term, freq in term_freq.items() if term in idf}
    return query_vector


def rank_by_tf(query, documents):
    query_terms = query.split()
    rankings = []
    for doc_id, doc_text in documents.items():
        doc_terms = doc_text.split()
        score = sum(doc_terms.count(term) for term in query_terms)
        if(score!=0):
            rankings.append((doc_id, score))
    return sorted(rankings, key=lambda x: x[1], reverse=True)


def rank_by_idf(query, documents):
    query_terms = query.split()
    doc_freqs = Counter()
    for doc_text in documents.values():
        doc_terms = set(doc_text.split())
        for term in doc_terms:
            if term in query_terms:
                doc_freqs[term] += 1
    rankings = []
    for doc_id, doc_text in documents.items():
        doc_terms = doc_text.split()
        score = sum(math.log(len(documents)/doc_freqs[term]) * doc_terms.count(term) for term in query_terms)
        if(score!=0):
            rankings.append((doc_id, score))
    return sorted(rankings, key=lambda x: x[1], reverse=True)


def rank_by_cosine(query, documents):
    query_terms = query.split()
    doc_vectors = {i: create_doc_vector(doc) for i, doc in enumerate(documents)}
    query_vector = create_query_vector(query_terms, documents)
    rankings = []
    for doc_id, doc_vector in doc_vectors.items():
        doc_length = math.sqrt(sum(value**2 for value in doc_vector.values()))
        query_length = math.sqrt(sum(value**2 for value in query_vector.values()))
        if doc_length > 0 and query_length > 0:
            score = sum(doc_vector.get(term, 0) * query_vector.get(term, 0) for term in query_terms) / (doc_length * query_length)
            if(score!=0):
                rankings.append((doc_id, score))
    return sorted(rankings, key=lambda x: x[1], reverse=True)



def rank_by_bm25(query, documents, k1=1.2, b=0.75):
    query_terms = query.split()
    doc_lengths = {doc_id: len(doc_text.split()) for doc_id, doc_text in documents.items()}
    avg_doc_len = sum(doc_lengths.values()) / len(documents)
    doc_freqs = {}
    for term in query_terms:
        doc_freqs[term] = sum(1 for doc_text in documents.values() if term in doc_text.split())
    rankings = []
    for doc_id, doc_text in documents.items():
        doc_terms = doc_text.split()
        score = 0
        for term in query_terms:
            tf = doc_terms.count(term)
            idf = math.log((len(documents) - doc_freqs[term] + 0.5) / (doc_freqs[term] + 0.5))
            score += idf * ((tf * (k1 + 1)) / (tf + k1 * (1 - b + b * (doc_lengths[doc_id] / avg_doc_len))))
        if(score!=0):
            rankings.append((doc_id, score))
    return sorted(rankings, key=lambda x: x[1], reverse=True)
```

### Relevance feedback, semantic matching, re-ranking of results, finding modified query :

```python
import numpy as np

def rocchio(query, relevant_docs, irrelevant_docs, alpha=1, beta=0.75, gamma=0.15):
    
    # Create vectors for the query, relevant documents, and irrelevant documents
    query_vector = create_query_vector(query, relevant_docs)
    relevant_vectors = [create_doc_vector(doc) for doc in relevant_docs]
    irrelevant_vectors = [create_doc_vector(doc) for doc in irrelevant_docs]

    # Calculate the centroid of the relevant documents
    relevant_centroid = np.mean(relevant_vectors, axis=0)

    # Calculate the centroid of the irrelevant documents
    irrelevant_centroid = np.mean(irrelevant_vectors, axis=0)

    # Apply the Rocchio algorithm to get the new query vector
    new_query_vector = alpha * np.array(list(query_vector.values())) + beta * relevant_centroid - gamma * irrelevant_centroid
    
    # Convert the new query vector back to a query string
    new_query_terms = []
    for i, term in enumerate(query_vector.keys()):
        new_query_terms.append((term, new_query_vector[i]))
    new_query_terms.sort(key=lambda x: x[1], reverse=True)
    new_query = ' '.join([term[0] for term in new_query_terms])
    
    return new_query
```
